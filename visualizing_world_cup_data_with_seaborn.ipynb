```python
#import codecademylib3_seaborn
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the world cup matches data:
df = pd.read_csv('WorldCupMatches.csv')
print(df.head(10))

# Adding a total goals column:
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df.head())

# Plotting a bar chart of goals in each year:
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.4)

f, ax = plt.subplots(figsize=(10, 25))
ax = sns.barplot(data=df, x='Year', y='Total Goals')
ax.set_title('Average Number Of Goals Scored In World Cup Matches By Year')
plt.show()

# Loading the goals dataset:
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())

# Plotting a boxplot:
sns.set_context('notebook', font_scale=1.25)
f, ax2 = plt.subplots(figsize=(12,7))
ax2 = sns.boxplot(data=df_goals, x='year', y='goals')
sns.set_palette('Spectral')
ax2.set_title('Range of Goals across each year')
plt.show()
```
