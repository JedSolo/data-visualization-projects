```python
import pandas as pd
import matplotlib.pyplot as plt

# load rankings data here:
wood_ranking = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
steel_ranking = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')

print(wood_ranking.head())
print(steel_ranking.head())

# write function to plot rankings over time for 1 roller coaster here:
def plot_line(name, data):
    year = data['Year of Rank'][data.Name == name]
    rank = data.Rank[data.Name == name]
    plt.plot(year, rank)
    ax = plt.subplot()
    ax.invert_yaxis()
    plt.title('Ranking Over Time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.show()

plot_line("El Toro", wood_ranking)


plt.clf()

# write function to plot rankings over time for 2 roller coasters here:


def plot_line_2(name1, name2, data):
    year_1= data['Year of Rank'][data.Name == name1]
    rank_1 = data.Rank[data.Name == name1]
    year_2= data['Year of Rank'][data.Name == name2]
    rank_2 = data.Rank[data.Name == name2]
    plt.plot(year_1, rank_1, label=name1)
    plt.plot(year_2, rank_2, label=name2)
    plt.legend()
    ax = plt.subplot()
    ax.invert_yaxis()
    plt.title('Ranking Over Time')
    plt.xlabel('Year')
    plt.ylabel('Rank')
    plt.show()

plot_line_2("El Toro", "Boulder Dash", wood_ranking)


plt.clf()


# write function to plot top n rankings over time here:

def plot_top_n(n, data):
    top_n_rankings = data[data['Rank'] <= n]
    
    for coaster in set(top_n_rankings['Name']):
        coaster_rankings = top_n_rankings[top_n_rankings['Name'] == coaster]
        plt.plot(coaster_rankings['Year of Rank'],coaster_rankings['Rank'],label=coaster)
        
    plt.legend()
    plt.show()
    

plot_top_n(5, wood_ranking)



plt.clf()
```

```python
import seaborn as sns
# load roller coaster data here:

roller_coasters = pd.read_csv('roller_coasters.csv')
print(roller_coasters.head())

# write function to plot histogram of column values here:

def plot_hist(column, data):
    values = data[column]
    plt.hist(values)
    plt.show()


plot_hist('speed', roller_coasters)



plt.clf()

# write function to plot inversions by coaster at a park here:

def plot_bar(name, data):
    park_coasters = data[data['park'] == name]
    park_coasters = park_coasters.sort_values('num_inversions', ascending=False)
    coaster_names = park_coasters['name']
    number_inversions = park_coasters['num_inversions']
    
    plt.bar(range(len(coaster_names)), number_inversions)
    ax = plt.subplot()
    ax.set_xticks(range(len(coaster_names)))
    ax.set_xticklabels(coaster_names, rotation=90)
    plt.show()

plot_bar('Bobbejaanland', roller_coasters)

    
    
    
plt.clf()

# write function to plot pie chart of operating status here:
def plot_pie(data):
    op_coasters = data[data.status == 'status.operating']
    cl_coasters = data[data.status == 'status.closed.definitely']
    list1 = [len(op_coasters), len(cl_coasters)]
    labels = ['Operating Roller Coasters', 'Closed Roller Coasters']
    plt.pie(list1, labels=labels, autopct='%d%%')
    plt.axis('equal')
    plt.show()

plot_pie(roller_coasters)


plt.clf()

# write function to create scatter plot of any two numeric columns here:

def plot_scatter(column1, column2, data):
    plt.scatter(x=data[column1], y=data[column2])
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()

plot_scatter('speed', 'height', roller_coasters)


plt.clf()

# Most popular roller coaster seating type:
counts = roller_coasters.seating_type.value_counts()
modal_count = counts.index[0]
print(modal_count)
```
